# **Ejercicio 2**
Asumir ahora que las funciones f1 y f2 representan las siguientes señales:
> * f1(x) = [1, 2, 3, 9]  a1 = 5
> * f2(x) = [13, 0, -1, 2]   a2 = 2

* Comprobar la linealidad si H es un operador 𝚺
* Comprobar la linealidad si H es un operador max

```python
import numpy as np
f1 = np.array([1,2,3,9])
f2 = np.array([13, 0, -1, 2])
a1 = 5
a2 = 2
print("f1 = ", f1)
print("f2 = ", f2)

ec1 = np.sum(a1*f1 + a2*f2)
ec2 = np.sum(a1*f1) + np.sum(a2*f2)

print("(a) Sumatora: ")
print(ec1, "=", ec2)  #Las ecuaciones son iguales

ec1 = np.max(a1*f1 + a2*f2)
ec2 = np.max(a1*f1) + np.max(a2*f2)

print("(b) Función Max: ")
print(ec1, "!=", ec2)   #Las ecuaciones no son iguales
```

# **Ejercicio 3**
Realizar un programa en python que calcule diferentes
operaciones matriciales de las imágenes mostradas.
Se pueden utilizar paqueterías como opencv, numpy o imutils
* Negativo de una imagen escala de grises
* Negativo de una imagen a color
* Transpuesta de la imagen
* Multiplicación por un escalar
* Doblado horizontal o vertical
* Rotación en 20 grados
* Escalamiento 



```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
import imutils

img1 = cv2.imread("zebra.png", 0) 
img2 = cv2.imread("flowers.png", 1)
img3 = cv2.imread("parrots.png", 1)

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)


def printPlot(IN, OUT, output_title):
  fig, (ax1, ax2) = plt.subplots(1, 2)
  ax1.imshow(IN)
  ax1.set_title("Original image")
  ax2.imshow(OUT)
  ax2.set_title(output_title)

#Negatives
INV1 = 255 - img1
INV2 = 255 - img2
printPlot(img1, INV1, "Negativo de una imagen a escala de grises")
printPlot(img2, INV2, "Negativo de una imagen a color")

#Transpose

T = cv2.transpose(img3)
printPlot(img3, T, "Transpuesta de la imagen")

#Scalar multiplicaction
S = img3*3
printPlot(img3, S, "Multiplicación escalar (3)")

#Horizontal fold
H = cv2.flip(img2, 1)
printPlot(img2, H, "Doblado horizontal")

#20 Rotation
R = imutils.rotate(img3, 20)
printPlot(img3, R, "Rotación 20 grados")

#Resize
E = cv2.resize(img2, (300, 400), interpolation=cv2.INTER_LINEAR)
printPlot(img2, E, "Escalamiento")


```
