# **Ejercicio 1**
En un programa en Python (NumPy)
calcula los valores CMYK y HSI de la
imagen original RGB mediante las
ecuaciones de equivalencias.


```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
import math

img = cv2.imread("strawberries_coffee_full_color.png")
B, G, R = cv2.split(img)

Rp = R/255
Gp = G/255
Bp = B/255


####### CMYK ##########

#### Metodo #1 #####

Y, M, C = 1-B, 1-G, 1-R

# K = 1 - cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
K = np.minimum(np.minimum(C, M), Y)

#### Metodo #2 #####

#K = 1-(np.maximum(np.maximum(R, G), B)/255)
# C = (1 - Rp - K)/(1-K)
# M = (1 - Gp - K)/(1-K)
# Y = (1 - Bp - K)/(1-K)

#### HSI ######
I = (1/3)*(Rp + Gp + Bp)

S = 1 - (3/(Bp + Gp + Rp))*np.minimum(np.minimum(Rp, Gp), Bp)  #255 or -1????

th_num = (1/2)*((Rp-Gp) + (Rp-Bp))
th_den = ((Rp-Gp)**2 + (Rp-Bp)*(Gp-Bp))**(1/2)
th = np.arccos(th_num/th_den)

H_mask = B <= G
# print(H_mask)
H = math.radians(360)-th
# H = 360-th    ??????
# print(H[H_mask])
H[H_mask] = th[H_mask]

##### CYMK #####
plt.figure()
f, axarr = plt.subplots(1,4)
f.set_size_inches(18.5, 10.5)
axarr[0].imshow(C, cmap='gray')
axarr[0].title.set_text('Cyan')
axarr[1].imshow(M, cmap='gray')
axarr[1].title.set_text('Magenta')
axarr[2].imshow(Y, cmap='gray')
axarr[2].title.set_text('Yellow')
axarr[3].imshow(K, cmap='gray')
axarr[3].title.set_text('K-Black')
##### RGB #####
plt.figure()
f, axarr = plt.subplots(1,3)
f.set_size_inches(18.5, 10.5)
axarr[0].imshow(R, cmap='gray')
axarr[0].title.set_text('Red')
axarr[1].imshow(G, cmap='gray')
axarr[1].title.set_text('Green')
axarr[2].imshow(B, cmap='gray')
axarr[2].title.set_text('Blue')
##### HSI #####
plt.figure()
f, axarr = plt.subplots(1,3)
f.set_size_inches(18.5, 10.5)
axarr[0].imshow(H, cmap='gray')
axarr[0].title.set_text('Hue')
axarr[1].imshow(S, cmap='gray')
axarr[1].title.set_text('Saturation')
axarr[2].imshow(I, cmap='gray')
axarr[2].title.set_text('Intensity')
```

# **Ejercicio 2**

Realiza un programa que permita separar una
imagen a color lena_saturada.png en distintos
canales RGB y HSI (usar funciones de opencv)

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

img = cv2.imread("lena_saturada.png")

B, G, R = cv2.split(img)

img_HSI = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)

H, S, I = cv2.split(img_HSI)

##### RGB #####
plt.figure()
f, axarr = plt.subplots(1,3)
f.set_size_inches(18.5, 10.5)
axarr[0].imshow(R, cmap='gray')
axarr[0].title.set_text('Red')
axarr[1].imshow(G, cmap='gray')
axarr[1].title.set_text('Green')
axarr[2].imshow(B, cmap='gray')
axarr[2].title.set_text('Blue')
##### HSI #####
plt.figure()
f, axarr = plt.subplots(1,3)
f.set_size_inches(18.5, 10.5)
axarr[0].imshow(H, cmap='gray')
axarr[0].title.set_text('Hue')
axarr[1].imshow(S, cmap='gray')
axarr[1].title.set_text('Saturation')
axarr[2].imshow(I, cmap='gray')
axarr[2].title.set_text('Intensity')
```
