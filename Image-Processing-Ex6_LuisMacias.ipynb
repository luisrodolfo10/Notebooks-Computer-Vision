## **Ejercicio 1**
● Dadas dos imágenes binarias A y B como se
muestran a continuación. Realizar las siguientes operaciones
booleanas:

1.   NOT(A)
2.   A AND B
3.   A OR B
4.   A XOR B
5.   [ NOT(A) ] AND B


```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt
img_A = cv2.imread('img_A.png', 0)/255
img_B = cv2.imread('img_B.png', 0)/255

img_1 = np.logical_not(img_A)
img_2 = np.logical_and(img_A, img_B)
img_3 = np.logical_or(img_A, img_B)
img_4 = np.logical_xor(img_A, img_B)
img_5 = np.logical_and(np.logical_not(img_A), img_B)

plt.figure()
f, axarr = plt.subplots(2,3)
f.set_size_inches(20, 12.5)
axarr[0, 0].imshow(img_A, cmap='gray')
axarr[0, 0].title.set_text("img A")
axarr[0, 1].imshow(img_1, cmap='gray')
axarr[0, 1].title.set_text("img 1")
axarr[0, 2].imshow(img_2, cmap='gray')
axarr[0, 2].title.set_text("img 2")
axarr[1, 0].imshow(img_3, cmap='gray')
axarr[1, 0].title.set_text("img 3")
axarr[1, 1].imshow(img_4, cmap='gray')
axarr[1, 1].title.set_text("img 4")
axarr[1, 2].imshow(img_5, cmap='gray')
axarr[1, 2].title.set_text("img 5")
```

## **Ejercicio 2**
Realiza un programa en Python que permita realizar la
dilatación y la erosión de una imagen donuts.png

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# Reading the input image
img = cv2.imread('donuts.png', 0)
 
kernel_n = 9 #@param {type:"slider", min:1, max:21, step:1}

# Taking a matrix of size 5 as the kernel
kernel = np.ones((kernel_n,kernel_n), np.uint8)
 
# The first parameter is the original image,
# kernel is the matrix with which image is
# convolved and third parameter is the number
# of iterations, which will determine how much
# you want to erode/dilate a given image.
img_erosion = cv2.erode(img, kernel, iterations=1)
img_dilation = cv2.dilate(img, kernel, iterations=1)

print("Kernel size: ", kernel_n)
plt.figure()
f, axarr = plt.subplots(1,3)
f.set_size_inches(20, 12.5)
axarr[0].imshow(img, cmap='gray')
axarr[0].title.set_text("Original")
axarr[1].imshow(img_erosion, cmap='gray')
axarr[1].title.set_text("Erosion")
axarr[2].imshow(img_dilation, cmap='gray')
axarr[2].title.set_text("Dilation")
```

## **Ejercicio 4**
Mediante las siguientes equivalencias, realiza un programa
en Python para calcular las operaciones de Opening y
Closing en la imagen

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# Reading the input image
img = cv2.imread('fingerprint.png', 0)
 
kernel_n = 3 #@param {type:"slider", min:1, max:21, step:1}

# Taking a matrix of size 5 as the kernel
kernel = np.ones((kernel_n,kernel_n), np.uint8)
 
# The first parameter is the original image,
# kernel is the matrix with which image is
# convolved and third parameter is the number
# of iterations, which will determine how much
# you want to erode/dilate a given image.
img_erosion = cv2.erode(img, kernel, iterations=1)
img_dilation = cv2.dilate(img, kernel, iterations=1)

img_opening = cv2.dilate(img_erosion, kernel, iterations=1)
img_closing = cv2.erode(img_dilation, kernel, iterations=1)

print("Kernel size: ", kernel_n)
plt.figure()
f, axarr = plt.subplots(1,3)
f.set_size_inches(20, 12.5)
axarr[0].imshow(img, cmap='gray')
axarr[0].title.set_text("Original")
axarr[1].imshow(img_opening, cmap='gray')
axarr[1].title.set_text("Opening")
axarr[2].imshow(img_closing, cmap='gray')
axarr[2].title.set_text("Closing")
```

## **Ejercicio 6**
De la imagen Lincoln_from_penny.png realiza la operación de extracción de
bordes mediante la ecuación:image.png

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt

# Reading the input image
img = cv2.imread('Lincoln_from_penny.png', 0)
 
kernel_n = 3 #@param {type:"slider", min:1, max:21, step:1}

# Taking a matrix of size 5 as the kernel
kernel = np.ones((kernel_n,kernel_n), np.uint8)

img_erosion = cv2.erode(img, kernel, iterations=1)

img_edges = img - img_erosion

print("Kernel size: ", kernel_n)
plt.figure()
f, axarr = plt.subplots(1,3)
f.set_size_inches(20, 12.5)
axarr[0].imshow(img, cmap='gray')
axarr[0].title.set_text("Original")
axarr[1].imshow(img_erosion, cmap='gray')
axarr[1].title.set_text("Erosion")
axarr[2].imshow(img_edges, cmap='gray')
axarr[2].title.set_text("Edges_detection")
```
