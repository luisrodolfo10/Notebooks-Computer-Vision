# **Ejemplo 2**
Realiza un programa en python que aplique la
corrección gamma de la imagen aérea que se
muestra en la imagen original mediante la
ecuación
s = c ( r + ε ) ^ y


```python
import numpy as np
import cv2
from google.colab.patches import cv2_imshow
import matplotlib.pyplot as plt


img = cv2.imread('aerial.png')
#img = np.array(img).astype('uint8')
c = 1
g = 3
p_img = c*(img + 10**-6)**g
p_img = p_img/np.max(p_img)
#print(p_img)

plt.imshow(p_img)
```

# **Ejemplo 4**
Realizar un programa en Python que imprima el histograma de una imagen a
color, por cada canal RGB por separado, tal como se muestra a continuación

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('Lenna.png')
cv2_imshow(img)

color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```
